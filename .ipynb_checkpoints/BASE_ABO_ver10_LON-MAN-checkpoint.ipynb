***Solution Name:Audience Bid Optmizer***

***Overview***
Audience Bid Optimizer provides machine learning based propensity score to convert and split cookie id based on these score.
Trading desk team will modify bid price with these audience segments.

***USP(Unique Sales Point)***
    1.Bespoke machine learning methodology and parameters for each client
    2.Huge and cleaned training data based on google placement category
    3.Easy to replicate because it based on google DT data .
    
   
***Data Source***
    cookie id: google dcm user id
    training data: conversion flag and DBM advertising placement category.(https://storage.googleapis.com/adx-rtb-dictionaries/publisher-verticals.txt)


***Processes***
    1.data extraction
    2.training and evaluating ML methodologies
    3.Determining ML method and parameters
    4.Scoring
    5.Uploading segments

***Total time of running code***
   Approximately 4-5 hours

***Necessary inputs***
Market information in big query
   -project id
   -dataset name

Audience extraction options
   -impression frequency:you set high fq, data per user increase and your would get more robust result but reduce number of users
   -time period.

ID to specify your client conversion and client themselves.
   -audience id
   -account id
   
Audience ids to flight your audience in DBM and location of GCS your client team use.
   -gcs bucket name
   -specific folder location you set 

***Consideration***
If your market use different name convention rule about table or DT transfer, the code need to be amended.
